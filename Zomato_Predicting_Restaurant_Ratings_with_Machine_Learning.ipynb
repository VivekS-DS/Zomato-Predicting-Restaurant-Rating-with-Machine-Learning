```python
# function for creating box plots for numeric columns to check for outliers

def check_outlier_boxplot(numeric_col):

  # Calculate the number of rows and columns for the subplot grid
  n_col = len(numeric_col)

  n_rows = (n_col + 2) // 3     # Create 3 columns per row

  # Set the figure size based on the number of rows
  fig, axs = plt.subplots(nrows=n_rows,ncols=3,figsize=(15,8*n_rows))
  axs =axs.flatten()

  # Iterate through the numeric columns and create box plots
  for i, var in enumerate(numeric_col):
    sns.boxplot(y=hotel[var],ax=axs[i]) # Create a box plot for the variable
    axs[i].set_title(var)        # Set the subplot title to the variable name

  # If there are more subplots than numeric columns, remove the excess subplots
  if n_col < len(axs):
    for i in range(n_col,len(axs)):
      fig.delaxes(axs[i])

  fig.tight_layout() # Adjust the subplot layout for readability
  plt.show()

# def check_outlier_boxplot(numeric_col):usage:
check_outlier_boxplot(col_numeric)
```

```python
# Create count plots for categorical variables in the 'hotel' DataFrame
# This function performs univariate analysis by generating count plots
  #for each categorical column.

def cat_countplot(categoric):  # categoric Univariable analysis ==> Countplot
  # Calculate the number of categorical columns and rows for subplots
  num_col = len(categoric)
  num_rows = (num_col + 2)//2

  # Create a figure with subplots based on the number of categorical columns
  fig, axs = plt.subplots(nrows=num_col,ncols=2,figsize=(15,10*num_rows))
  axs =axs.flatten()

  # Iterate through each categorical variable
  for i, var in enumerate(categoric):
    # Get the top 5 most frequent values for the current variable
    top_values =hotel[var].value_counts().nlargest(5).index
    # Filter the 'hotel' DataFrame to include only rows with the top values
    filtered_hotel =hotel[hotel[var].isin(top_values)]
    # Create a count plot for the current variable
    plots=sns.countplot(x=var,data=filtered_hotel,ax=axs[i])
    axs[i].set_title(var)
    axs[i].tick_params(axis='x',rotation=90)

  # Annotate the bars in the count plot with their respective counts
    for bar in plots.patches:
       plots.annotate(format(bar.get_height(), '.0f'),
                      (bar.get_x() + bar.get_width() / 2,
                      bar.get_height()), ha='center', va='center',
                      size=8, xytext=(0, 8),
                      textcoords='offset points')

  # Remove any extra empty subplots
  if num_col < len(axs):
    for i in range(num_col,len(axs)):
      fig.delaxes(axs[i])

  # Adjust spacing between subplots for better readability
  fig.tight_layout()

  # Display the count plots
  plt.show()
  return
```

# **Machine Learning**

Preprocessing data for ML

```python
# Import necessary libraries for machine learning
from sklearn.model_selection import train_test_split

#from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
hotel.name.nunique()
```

```python
# Drop the 'name' column
#hotel.drop('name',axis=1, inplace=True)
```

```python
# Map categorical values in the 'rest_rating' column to numerical values
hotel['rest_rating'] = hotel['rest_rating'].map({'Low':0,'Medium':1,'High':2})
hotel['rest_rating']
```

```python
# Create dummy variables for categorical columns
#and drop the first category to avoid multicollinearity
new_data = pd.get_dummies(hotel,drop_first=True)
```

```python
# Get a list of column names from the 'new_data' DataFrame
columns_list = list(new_data.columns)
```

```python
# Create a list of features by subtracting the 'rest_rating' column
#from the list of all columns
features = list(set(columns_list)-set(['rest_rating']))
```

```python
# Extract the target variable 'rest_rating' and store it as a NumPy array 'y'
y = new_data['rest_rating'].values
print(y)
```

```python
# Extract the feature variables defined in 'features' and store them as a NumPy array 'x'
x = new_data[features].values
print(x)
```

```python
# Split the dataset into training and testing sets
# - 'x' contains the feature variables
# - 'y' contains the target variable
# - The 'test_size' parameter specifies the proportion of the dataset
    #to include in the test split (here, 20%)
# - The 'random_state' parameter ensures reproducibility of the split
# - The 'stratify' parameter preserves the class distribution of the
    #target variable in the split

train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.2,
                                                    random_state=0, stratify=y)
```

```python
train_x.shape, train_y.shape
```

```python
test_x.shape, test_y.shape
```

## **Decision Tree Classifier**

```python
# Import the DecisionTreeClassifier class from scikit-learn
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

Hyperparameter Tuning

```python
# Import the GridSearchCV class from scikit-learn
from sklearn.model_selection import GridSearchCV
```

```python
# Hyperparameters and the range of values to search
def GridSearchCV_dt():
  param_grid = {'criterion': ['gini', 'entropy'],
                'max_depth': [None, 5, 10, 15],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]
                }

  # GridSearchCV for hyperparameter tuning with a Decision Tree Classifier
  grid_search = GridSearchCV(estimator=DecisionTreeClassifier(),
                            param_grid=param_grid, cv=5,
                            scoring='accuracy')
  grid_search.fit(train_x,train_y)

  return print(grid_search.best_params_)

#GridSearchCV_dt()
```

```python
# Create an instance of the Decision Tree Classifier
dt = DecisionTreeClassifier(criterion= 'gini', max_depth= None,
                            min_samples_leaf= 1, min_samples_split= 2)

# Fit (train) the Decision Tree Classifier using the training data
dt.fit(train_x,train_y)

# Training accuracy
print('Decision Tree Accuracy (Training)',dt.score(train_x,train_y))

# Use the trained Decision Tree Classifier to make predictions on the test data
pred_dt =dt.predict(test_x)

# Prediction accuracy
print('Decision Tree  Accuracy (Prediction)',accuracy_score(test_y, pred_dt))
```

## **Random Forest Classifier**

```python
# Import the Random Forest Classifier class from scikit-learn
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
# Hyperparameters and the range of values to search
def GridSearchCV_rfc():
  param_grid = {'criterion': ['gini', 'entropy'],
                'max_depth': [None, 5, 10, 15],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]
                }

  grid_search = GridSearchCV(estimator=rfc,
                            param_grid=param_grid, cv=5,
                            scoring='accuracy')

  grid_search.fit(train_x,train_y)

  return print(grid_search.best_params_)

#GridSearchCV_rfc()
```

```python
rfc = RandomForestClassifier(criterion= 'gini', max_depth= None,
                             min_samples_leaf= 1, min_samples_split= 2)

rfc.fit(train_x,train_y)

print('Random Forest Classifier Training Accuracy',rfc.score(train_x,train_y))

pred_rfc =rfc.predict(test_x)

# Prediction accuracy
print('Random Forest Classifier Prediction Accuracy',accuracy_score(test_y, pred_rfc))
```

## **KNN Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier

#KNN Classifier
knn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV

def GridSearchCV_knn():
  param_grid = {'n_neighbors': [1,2], 'weights': ['uniform', 'distance'],
                'algorithm':['ball_tree', 'kd_tree', 'brute']}

  grid_search = GridSearchCV(estimator=knn,
                            param_grid=param_grid, cv=5,
                            scoring='accuracy')

  grid_search.fit(train_x,train_y)

  return print(grid_search.best_params_)

GridSearchCV_knn()
```
